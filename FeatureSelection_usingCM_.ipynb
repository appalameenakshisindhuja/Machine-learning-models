```python
# Select rows 0 through 4 and columns 0 through 2
training=df.iloc[:, 0:6]
training.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
training= StandardScaler().fit_transform(training)
training
```

```python
import pandas as pd
df=pd.DataFrame(training)
df.describe()
```

```python
cor_matrix=df.corr().abs()
cor_matrix
```

```python
res=pd.DataFrame(cor_matrix)
cols=['feature1','feature2','feature3','feature4','feature5','feature6']
res.columns = cols
res.index = cols
res
```

```python
import numpy as np
upper = res.where(np.triu(np.ones(res.shape), k=1).astype(bool))
upper
```

```python
res1=upper.unstack()
res2=res1.dropna()
res2
```

```python
upper_int = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k=1).astype(bool))
upper_int.unstack().dropna().sort_values(ascending=True)
```

```python
to_drop = [column for column in upper_int.columns if any(upper_int[column] > 0.95)]
df.drop(to_drop, axis=1, inplace=True)
print("Dropped columns:", to_drop)
display(df.head())
```
